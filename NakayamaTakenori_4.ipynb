```python
import numpy as np 
import scipy.linalg
import matplotlib.pyplot as plt 

f = lambda x: x**2 + 2 * x + 2

np.random.seed(314)
x1 = np.arange(-4,5)
y = f(x1)
noise =  np.random.normal(scale=3, size=len(x1))

n = 3
X1 = np.zeros((len(x1), n))
X1[:, 0] = 1
X1[:, 1] = x1
X1[:, 2] = x1 ** 2
(param1, residuals, rank, s) = scipy.linalg.lstsq(X1, y+noise)

plt.title('least square')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x1, y + noise, color ='blue')
plt.plot(x1, param1[0] * X1[:, 0] + param1[1] * X1[:, 1]+ param1[2]*X1[:, 2], color = 'red')
```

```python
import numpy as np 
import scipy.linalg
import matplotlib.pyplot as plt 

f = lambda x: x**2 + 2 * x + 2

np.random.seed(314)
x2= np.arange(-4,4+0.01,0.01)
y = f(x2)
noise =  np.random.normal(scale=3, size=len(x2))

n = 3
X2 = np.zeros((len(x2), n))
X2[:, 0] = 1
X2[:, 1] = x2
X2[:, 2] = x2 ** 2
(param2, residuals, rank, s) = scipy.linalg.lstsq(X2, y+noise)

plt.title('least square')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x2, y + noise, color ='blue')
plt.plot(x2, param2[0] * X2[:, 0] + param2[1] * X2[:, 1]+ param2[2]*X2[:, 2], color = 'red')
```
